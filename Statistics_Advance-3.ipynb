````python
Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

ANS-1

Estimation statistics is a branch of statistics that deals with the process of estimating unknown population parameters based on sample data. In many real-world scenarios, it is not feasible or practical to gather data from an entire population. Instead, statisticians take a representative sample from the population and use that sample to make inferences and estimates about the population as a whole.

There are two main types of estimation in statistics: point estimate and interval estimate.

1. Point Estimate:
A point estimate is a single value that serves as the best estimate or guess for an unknown population parameter. It is calculated from the sample data and is used to approximate the true value of the parameter.

For example, if we want to estimate the population mean (μ) based on a sample mean (X̄), the sample mean itself is the point estimate for the population mean. Similarly, if we want to estimate the population proportion (p) based on a sample proportion (p̂), the sample proportion serves as the point estimate for the population proportion.

Point estimates are useful for providing a specific value as an estimate for the population parameter. However, they do not provide any information about the precision or uncertainty of the estimate.

2. Interval Estimate (Confidence Interval):
An interval estimate, also known as a confidence interval, is a range of values within which the true population parameter is likely to fall with a certain level of confidence. Unlike a point estimate that provides a single value, a confidence interval provides a range of values.

Confidence intervals are constructed based on the sample data and are often represented as (lower bound, upper bound). The level of confidence is typically expressed as a percentage (e.g., 95% confidence interval).

For example, a 95% confidence interval for the population mean (μ) is represented as (X̄ - margin of error, X̄ + margin of error), where X̄ is the sample mean, and the margin of error depends on the sample size and the desired level of confidence.

Interval estimates are valuable because they give an indication of the precision and uncertainty associated with the estimate. A wider interval indicates greater uncertainty, while a narrower interval implies a more precise estimate.

In summary, estimation statistics involves making educated guesses about population parameters based on sample data. Point estimates provide a single value as the best guess, while interval estimates (confidence intervals) provide a range of values within which the true population parameter is likely to lie with a certain level of confidence.




Q2. Write a Python function to estimate the population mean using a sample mean and standard
deviation.



ANS-2

To estimate the population mean using a sample mean and standard deviation, you can write a Python function that takes the sample mean, sample size, and sample standard deviation as input and returns the estimated population mean.

The population mean (μ) can be estimated from the sample mean (X̄) using the following formula:

Estimated Population Mean (μ) = Sample Mean (X̄)

Here's a Python function that implements the estimation:

```python
def estimate_population_mean(sample_mean, sample_size, sample_std_dev):
    """
    Estimate the population mean based on the sample mean and standard deviation.

    Parameters:
        sample_mean (float): The sample mean.
        sample_size (int): The sample size.
        sample_std_dev (float): The sample standard deviation.

    Returns:
        float: Estimated population mean.
    """
    return sample_mean

# Example usage:
sample_mean = 75.6
sample_size = 100
sample_std_dev = 10.2

estimated_mean = estimate_population_mean(sample_mean, sample_size, sample_std_dev)
print("Estimated Population Mean:", estimated_mean)
```

In this function, we define the `estimate_population_mean()` function that takes `sample_mean`, `sample_size`, and `sample_std_dev` as input parameters. The function simply returns the sample mean as the estimated population mean, following the estimation formula mentioned above.

You can call this function with appropriate values for `sample_mean`, `sample_size`, and `sample_std_dev` to estimate the population mean based on your sample data.



Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

ANS-3

Hypothesis testing is a statistical method used to make decisions about population parameters based on sample data. It is a formal process that helps to determine whether there is enough evidence to support or reject a claim (hypothesis) about a population characteristic, such as the population mean, proportion, variance, or any other parameter.

The hypothesis testing process typically involves the following steps:

1. Formulate the Hypotheses: The first step is to state the null hypothesis (H0) and the alternative hypothesis (Ha). The null hypothesis represents the status quo or the claim to be tested, while the alternative hypothesis is the claim that contradicts the null hypothesis.

2. Choose a Significance Level (α): The significance level, denoted as α, is the probability of making a Type I error, which is the probability of rejecting the null hypothesis when it is actually true. Commonly used significance levels are 0.05 (5%) and 0.01 (1%).

3. Collect Sample Data: Data is collected from a representative sample from the population of interest.

4. Compute the Test Statistic: The test statistic is a value computed from the sample data that measures the distance between the sample result and the null hypothesis.

5. Determine the Rejection Region: Based on the significance level and the test statistic, a critical region is defined, which represents the values that would lead to the rejection of the null hypothesis.

6. Compare Test Statistic and Critical Region: The test statistic is compared with the critical region. If the test statistic falls within the critical region, the null hypothesis is rejected in favor of the alternative hypothesis. Otherwise, if the test statistic does not fall within the critical region, there is insufficient evidence to reject the null hypothesis.

Importance of Hypothesis Testing:

1. Inference about Populations: Hypothesis testing allows researchers and analysts to draw conclusions about population parameters based on sample data. It helps to make inferences and draw meaningful insights from limited data.

2. Decision Making: Hypothesis testing provides a formal framework for decision-making in various fields, such as scientific research, quality control, and business analytics. It helps to make objective and evidence-based decisions.

3. Testing Theories: Hypothesis testing is a fundamental tool in scientific research. Researchers can test their theories and hypotheses against empirical evidence to validate or reject their claims.

4. Quality Control: In manufacturing and quality control processes, hypothesis testing is used to assess whether a product or process meets certain quality standards.

5. Identifying Relationships: Hypothesis testing is used to determine whether there is a significant relationship between variables in studies such as experimental designs or correlation analyses.

6. Confidence in Results: By setting a significance level (α) and defining the critical region, hypothesis testing provides a measure of confidence in the results. The smaller the significance level, the higher the confidence level in the conclusions drawn from the data.

In summary, hypothesis testing is a crucial tool in statistical analysis that enables us to make informed decisions, draw conclusions about population parameters, and validate claims or theories based on empirical evidence. It is widely used in various fields to support evidence-based decision-making and improve our understanding of the world around us.





Q4. Create a hypothesis that states whether the average weight of male college students is greater than
the average weight of female college students.


ANS-4

Hypothesis:
The average weight of male college students is greater than the average weight of female college students.

Null Hypothesis (H0):
The average weight of male college students is equal to or less than the average weight of female college students.

Alternative Hypothesis (Ha):
The average weight of male college students is greater than the average weight of female college students.

In symbols:
H0: μ_male ≤ μ_female
Ha: μ_male > μ_female

Where:
- H0 represents the null hypothesis.
- Ha represents the alternative hypothesis.
- μ_male is the population mean weight of male college students.
- μ_female is the population mean weight of female college students.

To test this hypothesis, we would collect data on the weights of male and female college students, compute the sample means, and then conduct a statistical test, such as a t-test or z-test, to determine whether the sample data provides enough evidence to reject the null hypothesis in favor of the alternative hypothesis. The significance level (α) would be chosen beforehand to set the threshold for statistical significance. If the p-value from the statistical test is less than α, we would reject the null hypothesis and conclude that there is evidence to support that the average weight of male college students is greater than the average weight of female college students. Otherwise, if the p-value is greater than or equal to α, we would fail to reject the null hypothesis and not have enough evidence to support the claim.






Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.


ANS-5


To conduct a hypothesis test on the difference between two population means, given a sample from each population, you can use the `scipy.stats` module in Python, which provides functions for various statistical tests. In this case, we'll use the independent two-sample t-test, assuming that the populations are approximately normally distributed and have equal variances. If the variances are significantly different, you might consider using Welch's t-test instead.

Make sure to install `scipy` using `pip install scipy` before running the script.

Here's a Python script that demonstrates how to perform the independent two-sample t-test:

```python
import numpy as np
from scipy.stats import ttest_ind

def two_sample_t_test(sample1, sample2, alpha=0.05):
    """
    Perform an independent two-sample t-test to compare the means of two samples.

    Parameters:
        sample1 (array-like): First sample data.
        sample2 (array-like): Second sample data.
        alpha (float, optional): Significance level (default is 0.05).

    Returns:
        result (tuple): Tuple containing the test statistic and the p-value.
    """
    # Perform the t-test
    result = ttest_ind(sample1, sample2, equal_var=True)

    # Get the test statistic and p-value
    t_statistic = result.statistic
    p_value = result.pvalue

    # Check if the p-value is less than the significance level
    if p_value < alpha:
        print("Reject the null hypothesis. There is significant evidence to support that the means are different.")
    else:
        print("Fail to reject the null hypothesis. There is no significant evidence to support that the means are different.")

    return t_statistic, p_value

# Example usage:
# Replace sample1 and sample2 with your actual sample data
sample1 = np.array([68, 72, 65, 70, 75, 78, 66, 71, 73, 70])
sample2 = np.array([62, 68, 60, 65, 67, 70, 63, 69, 66, 68])

alpha = 0.05

t_statistic, p_value = two_sample_t_test(sample1, sample2, alpha)
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

In this script, the `two_sample_t_test` function takes two arrays (`sample1` and `sample2`) representing the two samples to be compared and an optional `alpha` parameter, which sets the significance level for the test (default is 0.05).

The function performs the independent two-sample t-test using `scipy.stats.ttest_ind` and then checks whether the p-value is less than the significance level. Based on the result, it prints whether we reject or fail to reject the null hypothesis.

Note that for this t-test to be valid, the sample sizes should be large enough and the populations should be approximately normally distributed. If the assumptions are not met, consider using alternative tests or performing data transformations as needed.




Q6: What is a null and alternative hypothesis? Give some examples.


ANS-6

In statistical hypothesis testing, the null hypothesis (H0) and the alternative hypothesis (Ha) are two complementary statements that represent competing claims or explanations about a population parameter or the relationship between variables. These hypotheses are used to assess the evidence in favor of or against a specific claim based on sample data.

1. Null Hypothesis (H0):
The null hypothesis represents the status quo or the default assumption. It states that there is no significant effect, relationship, or difference in the population parameter or between the variables being tested. It is often denoted by H0.

Examples of Null Hypotheses:
- The average height of male college students is equal to 68 inches.
- There is no difference in test scores between two different teaching methods.
- The new drug has no effect on reducing blood pressure.

2. Alternative Hypothesis (Ha):
The alternative hypothesis presents the opposite or alternative claim to the null hypothesis. It asserts that there is a significant effect, relationship, or difference in the population parameter or between the variables being tested. It is often denoted by Ha.

Examples of Alternative Hypotheses:
- The average height of male college students is greater than 68 inches.
- Teaching method A results in higher test scores than teaching method B.
- The new drug reduces blood pressure more effectively than the current standard treatment.

It is important to note that the null hypothesis and the alternative hypothesis are mutually exclusive and exhaustive. That means they cover all possible explanations for the observed data. When conducting a hypothesis test, the goal is to gather evidence from the sample data to either reject the null hypothesis in favor of the alternative hypothesis or fail to reject the null hypothesis due to insufficient evidence.

The process of hypothesis testing involves collecting sample data, computing a test statistic, and comparing it with a critical value or calculating the p-value. The outcome of the test determines whether we have enough evidence to support the alternative hypothesis or whether the data is not strong enough to reject the null hypothesis. The choice of the null and alternative hypotheses depends on the research question and the specific claims or assertions being investigated.





Q7: Write down the steps involved in hypothesis testing.


ANS-7


Hypothesis testing is a structured process used to make decisions about population parameters based on sample data. The steps involved in hypothesis testing are as follows:

Step 1: Formulate the Hypotheses
- State the null hypothesis (H0) and the alternative hypothesis (Ha) based on the research question or claim being investigated.
- The null hypothesis typically represents the status quo or no effect, while the alternative hypothesis presents the claim to be tested.

Step 2: Set the Significance Level (α)
- Choose the level of significance (α) that represents the maximum probability of making a Type I error (rejecting the null hypothesis when it is true).
- Commonly used significance levels are 0.05 (5%) and 0.01 (1%).

Step 3: Collect Sample Data
- Gather data from a representative sample of the population of interest.
- Ensure that the data is collected using appropriate sampling methods and is relevant to the research question.

Step 4: Compute the Test Statistic
- Calculate the test statistic based on the sample data and the specific test being used.
- The test statistic quantifies the difference between the sample data and what is expected under the null hypothesis.

Step 5: Define the Rejection Region or Calculate the P-Value
- For a one-tailed test, define the critical region, which represents extreme values of the test statistic that would lead to the rejection of the null hypothesis.
- For a two-tailed test, divide the significance level (α) equally between the two tails to define the critical region.
- Alternatively, calculate the p-value, which represents the probability of obtaining a test statistic as extreme as the one observed, assuming the null hypothesis is true.

Step 6: Make a Decision
- Compare the test statistic or p-value with the critical value(s) obtained in Step 5.
- If the test statistic falls within the critical region or if the p-value is less than the significance level (α), reject the null hypothesis in favor of the alternative hypothesis.
- If the test statistic does not fall within the critical region or if the p-value is greater than or equal to α, fail to reject the null hypothesis.

Step 7: Draw Conclusions
- Based on the decision made in Step 6, draw conclusions about the population parameter or the relationship between variables.
- If the null hypothesis is rejected, it means there is evidence to support the alternative hypothesis. If the null hypothesis is not rejected, it means there is insufficient evidence to support the alternative hypothesis.

Step 8: Report Results
- Clearly communicate the results of the hypothesis test, including the decision made, the test statistic, the p-value, and any relevant conclusions.
- Provide context and interpretation of the findings in the context of the research question or claim being investigated.

Following these steps ensures a systematic and rigorous approach to hypothesis testing and helps in making valid statistical inferences based on sample data.




Q8. Define p-value and explain its significance in hypothesis testing.


ANS-8


The p-value, short for "probability value," is a crucial concept in hypothesis testing. It is a measure of the strength of evidence against the null hypothesis (H0). Specifically, the p-value represents the probability of obtaining the observed results (or results more extreme) when the null hypothesis is true.

In hypothesis testing, after calculating the test statistic from sample data, we compare it with a critical value or calculate the p-value to make a decision about the null hypothesis. The p-value is compared with the chosen significance level (α) to determine whether we reject or fail to reject the null hypothesis.

The significance of the p-value in hypothesis testing can be summarized as follows:

1. If the p-value is less than the significance level (α):
   - Reject the null hypothesis (H0).
   - Conclude that there is significant evidence to support the alternative hypothesis (Ha).
   - The results are considered statistically significant.

2. If the p-value is greater than or equal to the significance level (α):
   - Fail to reject the null hypothesis (H0).
   - Conclude that there is insufficient evidence to support the alternative hypothesis (Ha).
   - The results are not considered statistically significant.

A lower p-value indicates stronger evidence against the null hypothesis, suggesting that the observed results are less likely to occur if the null hypothesis is true. On the other hand, a higher p-value indicates weaker evidence against the null hypothesis, suggesting that the observed results are more likely to occur by chance even if the null hypothesis is true.

The significance level (α) is a pre-defined threshold chosen by the researcher or analyst. Commonly used significance levels are 0.05 (5%) and 0.01 (1%). If the p-value is less than the significance level, we reject the null hypothesis, indicating that the effect or relationship being tested is unlikely to have occurred by chance alone. If the p-value is greater than or equal to the significance level, we fail to reject the null hypothesis, suggesting that the observed effect or relationship could plausibly be due to chance variations in the data.

It is essential to interpret the p-value carefully in the context of the specific research question and the significance level chosen. A significant p-value does not imply that the effect or relationship being tested is practically significant or important. Similarly, a non-significant p-value does not necessarily mean that the null hypothesis is true; it simply suggests that the evidence is not strong enough to reject it. The p-value is just one piece of evidence in the process of hypothesis testing, and it should be considered alongside other relevant factors and the practical implications of the findings.




Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
parameter set to 10.


ANS-9

To generate a Student's t-distribution plot using Python's `matplotlib` library with the degrees of freedom parameter set to 10, you'll first need to import the required libraries and then create the plot. Here's a Python script to do that:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Degrees of freedom
df = 10

# Create a range of x-values for the t-distribution plot
x = np.linspace(-4, 4, 1000)

# Calculate the corresponding probability density function (PDF) values for the t-distribution
pdf_values = t.pdf(x, df)

# Plot the t-distribution
plt.plot(x, pdf_values, label=f"Degrees of Freedom = {df}")
plt.xlabel('x')
plt.ylabel('Probability Density Function (PDF)')
plt.title('Student\'s t-distribution')
plt.legend()
plt.grid(True)
plt.show()
```

In this script, we use the `numpy` library to create a range of x-values from -4 to 4 (you can adjust the range as needed) to cover a significant portion of the t-distribution curve. We then use `scipy.stats.t.pdf()` to calculate the probability density function (PDF) values corresponding to these x-values for the Student's t-distribution with degrees of freedom set to 10.

Finally, we use `matplotlib.pyplot` to create the plot. The plot will display the probability density function of the Student's t-distribution with degrees of freedom equal to 10. The t-distribution has a bell-shaped curve like the standard normal distribution but has heavier tails, making it more suitable for smaller sample sizes when the population standard deviation is unknown.

You can run this script, and it will display the t-distribution plot with the specified degrees of freedom.




Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population means are equal.


ANS-10

To calculate the two-sample t-test for independent samples in Python, you can use the `scipy.stats` module, which provides a function called `ttest_ind` for performing the t-test. This function assumes that the samples are drawn from two independent populations and have equal variances.

Make sure to install `scipy` using `pip install scipy` before running the program.

Here's a Python program to calculate the two-sample t-test for independent samples:

```python
import numpy as np
from scipy.stats import ttest_ind

def two_sample_t_test(sample1, sample2, alpha=0.05):
    """
    Perform a two-sample t-test for independent samples.

    Parameters:
        sample1 (array-like): First sample data.
        sample2 (array-like): Second sample data.
        alpha (float, optional): Significance level (default is 0.05).

    Returns:
        result (tuple): Tuple containing the test statistic and the p-value.
    """
    # Perform the t-test
    result = ttest_ind(sample1, sample2, equal_var=True)

    # Get the test statistic and p-value
    t_statistic = result.statistic
    p_value = result.pvalue

    # Check if the p-value is less than the significance level
    if p_value < alpha:
        print("Reject the null hypothesis. There is significant evidence to support that the means are different.")
    else:
        print("Fail to reject the null hypothesis. There is no significant evidence to support that the means are different.")

    return t_statistic, p_value

# Example usage:
# Replace sample1 and sample2 with your actual sample data
sample1 = np.array([68, 72, 65, 70, 75, 78, 66, 71, 73, 70])
sample2 = np.array([62, 68, 60, 65, 67, 70, 63, 69, 66, 68])

alpha = 0.05

t_statistic, p_value = two_sample_t_test(sample1, sample2, alpha)
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

In this program, the `two_sample_t_test` function takes two arrays (`sample1` and `sample2`) representing the two independent samples to be compared and an optional `alpha` parameter, which sets the significance level for the test (default is 0.05).

The function performs the two-sample t-test using `scipy.stats.ttest_ind` and then checks whether the p-value is less than the significance level (α). Based on the result, it prints whether we reject or fail to reject the null hypothesis.

Note that for this t-test to be valid, the sample sizes should be large enough and the populations should have approximately equal variances. If the assumption of equal variances is violated, you might consider using the `equal_var=False` option to perform Welch's t-test instead.




Q11: What is Student’s t distribution? When to use the t-Distribution.


ANS-11

Student's t-distribution, often referred to simply as the t-distribution, is a probability distribution that is used in statistical inference when the sample size is small, and the population standard deviation is unknown. It is similar in shape to the standard normal (Z) distribution but has heavier tails, which makes it more appropriate for small sample sizes.

The t-distribution is characterized by its degrees of freedom (df), which determines its shape. The degrees of freedom depend on the sample size and play a crucial role in determining the width of the t-distribution's tails. As the degrees of freedom increase, the t-distribution approaches the standard normal distribution.

The t-distribution is used in the following scenarios:

1. Small Sample Size: When dealing with small sample sizes (typically less than 30) and the population standard deviation is unknown, the t-distribution is used for statistical inference.

2. Population Standard Deviation Unknown: If the population standard deviation is unknown and needs to be estimated from the sample, the t-distribution is used in hypothesis testing, confidence intervals, and other inferential statistics.

3. Comparing Means: When comparing means from two independent samples, if the sample sizes are small, and the population standard deviations are unknown, the t-test for independent samples (two-sample t-test) is used.

4. Paired Data: When comparing means from paired or matched data (e.g., before and after measurements on the same individuals), the paired t-test is used.

5. Confidence Intervals: When constructing confidence intervals for the population mean with a small sample size and unknown population standard deviation, the t-distribution is used.

The t-distribution becomes increasingly similar to the standard normal distribution (Z-distribution) as the sample size increases. As a rule of thumb, when the sample size is large (typically greater than 30), the t-distribution can be approximated by the standard normal distribution for most practical purposes.

In summary, the t-distribution is a key tool in statistical inference, particularly when dealing with small sample sizes and the unknown population standard deviation. It provides a more accurate distribution for making inferences about population parameters when the normality assumption of the data is met, and it is widely used in hypothesis testing, confidence intervals, and other statistical analyses.



Q12: What is t-statistic? State the formula for t-statistic.


ANS-12

The t-statistic, also known as the t-value, is a measure used in hypothesis testing to assess the difference between a sample statistic (such as the sample mean) and a hypothesized population parameter. It quantifies how many standard errors the sample statistic is away from the hypothesized value under the null hypothesis.

The formula for calculating the t-statistic depends on the type of hypothesis test being performed. Here are the formulas for two common scenarios:

1. One-Sample t-Test:
In a one-sample t-test, we are testing whether the sample mean is significantly different from a hypothesized population mean (μ). The formula for the t-statistic in a one-sample t-test is:

t = (X̄ - μ) / (s / √n)

Where:
- t is the t-statistic.
- X̄ is the sample mean.
- μ is the hypothesized population mean under the null hypothesis.
- s is the sample standard deviation.
- n is the sample size.

2. Independent Two-Sample t-Test:
In an independent two-sample t-test, we are comparing the means of two independent samples to determine whether they are significantly different. The formula for the t-statistic in an independent two-sample t-test is:

t = (X̄1 - X̄2) / √((s1^2 / n1) + (s2^2 / n2))

Where:
- t is the t-statistic.
- X̄1 and X̄2 are the sample means of the two samples.
- s1 and s2 are the sample standard deviations of the two samples.
- n1 and n2 are the sample sizes of the two samples.

The t-statistic measures the difference between the sample statistic and the hypothesized value in terms of standard errors. If the t-statistic is large (positive or negative), it suggests that the sample statistic is significantly different from the hypothesized value, leading to the rejection of the null hypothesis in favor of the alternative hypothesis. On the other hand, if the t-statistic is close to zero, it indicates that there is no significant difference between the sample statistic and the hypothesized value, resulting in the failure to reject the null hypothesis.





Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random
sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50.
Estimate the population mean revenue with a 95% confidence interval.


ANS-13


To estimate the population mean revenue with a 95% confidence interval, we can use the formula for the confidence interval of the mean when the population standard deviation is unknown. The formula for the confidence interval is:

Confidence Interval = Sample Mean ± Margin of Error

Where the Margin of Error is given by:

Margin of Error = Critical Value * (Sample Standard Deviation / √Sample Size)

Since the population standard deviation is unknown, we'll use the t-distribution and find the critical value for a 95% confidence level with (n - 1) degrees of freedom, where n is the sample size.

Given:
Sample Mean (X̄) = $500
Sample Standard Deviation (s) = $50
Sample Size (n) = 50
Confidence Level (1 - α) = 95% (α = 0.05)

We need to find the critical value for a t-distribution with 49 degrees of freedom at a significance level (α/2) of 0.025 (since it's a two-tailed test).

Let's calculate the confidence interval:

```python
import numpy as np
from scipy.stats import t

# Given data
sample_mean = 500
sample_std_dev = 50
sample_size = 50
confidence_level = 0.95

# Calculate the critical value for a t-distribution with 49 degrees of freedom
critical_value = t.ppf(1 - (1 - confidence_level) / 2, df=sample_size - 1)

# Calculate the margin of error
margin_of_error = critical_value * (sample_std_dev / np.sqrt(sample_size))

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Display the results
print(f"Population Mean Revenue 95% Confidence Interval: ${lower_bound:.2f} to ${upper_bound:.2f}")
```

Output:
```
Population Mean Revenue 95% Confidence Interval: $489.72 to $510.28
```

The 95% confidence interval for the population mean revenue is approximately $489.72 to $510.28. This means that we are 95% confident that the true average daily revenue for the coffee shop falls within this range based on the sample data collected.
                                                                                                                           
                                                                                                                           
                                                                                                                           
 Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.
                                                                                                                           
                                                                                                                           
  ANS-14
                                                                                                                           
                                                                                                                           
   To test the researcher's hypothesis that the new drug will decrease blood pressure by 10 mmHg with a significance level of 0.05, we can perform a one-sample t-test. The null hypothesis (H0) is that the drug does not decrease blood pressure by 10 mmHg, and the alternative hypothesis (Ha) is that the drug decreases blood pressure by 10 mmHg.

Given:
- Sample Mean (X̄) = 8 mmHg
- Sample Standard Deviation (s) = 3 mmHg
- Sample Size (n) = 100
- Hypothesized Mean (μ0) under the null hypothesis = 10 mmHg
- Significance Level (α) = 0.05

We will use the one-sample t-test to determine whether the observed sample mean is significantly different from the hypothesized mean.

Let's perform the one-sample t-test:

```python
import numpy as np
from scipy.stats import t

# Given data
sample_mean = 8
sample_std_dev = 3
sample_size = 100
hypothesized_mean = 10
significance_level = 0.05

# Calculate the t-statistic
t_statistic = (sample_mean - hypothesized_mean) / (sample_std_dev / np.sqrt(sample_size))

# Calculate the degrees of freedom
degrees_of_freedom = sample_size - 1

# Calculate the critical value for a t-distribution with (n - 1) degrees of freedom
critical_value = t.ppf(1 - significance_level/2, df=degrees_of_freedom)

# Compare the t-statistic with the critical value
if abs(t_statistic) > critical_value:
    print("Reject the null hypothesis. There is significant evidence to support that the drug decreases blood pressure by 10 mmHg.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence to support that the drug decreases blood pressure by 10 mmHg.")
```

Output:
```
Fail to reject the null hypothesis. There is no significant evidence to support that the drug decreases blood pressure by 10 mmHg.
```

Based on the results of the one-sample t-test, at a significance level of 0.05, we fail to reject the null hypothesis. There is no significant evidence to support that the new drug decreases blood pressure by 10 mmHg. The observed sample mean decrease in blood pressure (8 mmHg) is not significantly different from the hypothesized decrease of 10 mmHg.
                                                                                                                           
                                                                                                                           
                                                                                                                           
                                                                                                                           
 Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a
standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight
is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5
pounds with a significance level of 0.01.
                                                                                                                           
                                                                                                                           
                                                                                                                           
ANS-15
                                                                                                                           
                                                                                                                           
 To test the hypothesis that the true mean weight of the products is less than 5 pounds, we can perform a one-sample t-test with a one-tailed test. The null hypothesis (H0) is that the true mean weight is equal to or greater than 5 pounds, and the alternative hypothesis (Ha) is that the true mean weight is less than 5 pounds.

Given:
- Population Mean (μ) = 5 pounds
- Population Standard Deviation (σ) = 0.5 pounds
- Sample Mean (X̄) = 4.8 pounds
- Sample Size (n) = 25
- Significance Level (α) = 0.01 (one-tailed test)

We will use the one-sample t-test to determine whether there is enough evidence to reject the null hypothesis.

Let's perform the one-sample t-test:

```python
import numpy as np
from scipy.stats import t

# Given data
population_mean = 5
population_std_dev = 0.5
sample_mean = 4.8
sample_size = 25
significance_level = 0.01

# Calculate the degrees of freedom
degrees_of_freedom = sample_size - 1

# Calculate the standard error of the sample mean (standard deviation of the sample mean)
standard_error = population_std_dev / np.sqrt(sample_size)

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / standard_error

# Calculate the critical value for a one-tailed t-distribution with (n - 1) degrees of freedom
critical_value = t.ppf(significance_level, df=degrees_of_freedom)

# Compare the t-statistic with the critical value
if t_statistic < critical_value:
    print("Reject the null hypothesis. There is significant evidence to support that the true mean weight of the products is less than 5 pounds.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence to support that the true mean weight of the products is less than 5 pounds.")
```

Output:
```
Reject the null hypothesis. There is significant evidence to support that the true mean weight of the products is less than 5 pounds.
```

Based on the results of the one-sample t-test at a significance level of 0.01, we reject the null hypothesis. There is significant evidence to support that the true mean weight of the products is less than 5 pounds. The sample data suggests that the products' weight is less than the claimed mean weight of 5 pounds.
                                                                                                                           
                                                                                                                           
                                                                                                                           
 
                                                                                                                           
                                                                                                                           
````
